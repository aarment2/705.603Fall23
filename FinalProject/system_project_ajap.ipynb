## SECTION 1: INTRODUCTION

Infertility is a growing problem worldwide. Infertility is defined as the inability to achieve pregnancy by either the male or the female reproductive systems of patients. In 2022, in the United States, an estimated 11% of women and 9% of men suffer from infertility. There are multiple reasons behind the growing trend of infertility, with our evolving lifestyles amongst the main contributing factors. A predominant consequence of modern life in couples seeking to start a family is the fact they are reaching this decision later in life, more so than previous generations. A male partner is more likely to experience infertility over the age 0f 40. Also, a woman's ovum's quality declines as she gets older, reducing the likelihood of a succesful pregnancy. Female ova also decline in quantity over the years. Other factors behind the rise in infertility include environmental factors, medications, smoking, and stress.

<img src="./img/img_01.jpg"  
     width="550" />

In-Vitro Fertilization (IVF) offers a viable solution to many couples facing infertility problems. IVF is the most widely used form of Assisted Reproductive Technology. Although the name might suggest this is a single procedure, in reality, IVF is a very complex process involving many steps that take several weeks or months to complete. At a high level, the IVF process starts with the female patient taking fertility medications for several months to induce ovum production. During this time, the female patient takes several blood tests and ultrasounds to monitor hormone levels and track ovum production. At a certain point, it is determined that the female patient has started to produce sufficient mature eggs. Following this, an IVF specialist retrieves eggs from the patient's body, which is a minor surgical procedure. Afterwards, the eggs retrieved are fertilized or inseminated with sperm cells from the male partner or donor. The inseminated eggs are stored in a special incubator for 3 to 5 days. During this time, the fertilized embryos go through several growth stages or milestones. After the incubation period, the doctor selects the best embryos based on experience and proceeds to transfer them to the female patient. Some patients may continue hormone treatment for several weeks after the embryo transfer. A succesful pregnancy occurs when at least one embryo attaches to the uterus and begins to develop properly.

     width="550" />

The success rate of In-Vitro Fertilization is an average of only 30%. As we have seen, IVF is a long and complicated process, and because of this, it has multiple points of failure. There are also several other factors, such as the age of the female patient, and lifestyle choices. Some important factors that may lead to the success or failure of an IVF procedure also include embryo quality, response of the ovarium, implantation issues, DNA and chromosomal abnormalities. Let us also consider that the average cost of an IVF cycle in the United States ranges somewhere between USD 15,000 and USD 30,000. With the low success rate, the majority of couples require going through the process several times to achieve a pregnancy, incurring in very high costs. Many other couples give up altogether without ever achieving a pregnancy. Another important cost factor is that the procedure is rarely covered by medical insurance, meaning patients need to pay full costs out of pocket. All of the context combined, with the costs and the low chances of succes, carry a large emotional toll for the couple seeking to get pregnant. Starting a family is a milestone in the life of many people, and having to go through an ART procedure for what otherwise occurs naturally, is extremely difficult, based on my personal experience supporting an IVF clinic.

It appears that, until recently two or three years ago, there had not been comprehensive research to understand the root cause or causes of IVF failure. Although some factors have long been established, such as the age of the mother and its correlation to embryo quality, those seemed to have a relatively minor impact on success. The ratio of success ranges from 25% to 40%, depending on the age of the mother. Then, 40% success still leaves a lot of room for improvement. Other factors, such as the environment, lifestyle of both parents, and even stress, have also been considered, but still did not move the needle much in terms of increasing chances of success. Recent studies are pointing to DNA replication issues as one of the main culprits.
